# Data Validations - Phase 2

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Environment ready")
```

```python
# Load Excel Data

file_path = "Fantasy-Lord-Rookie-ML-Dataset.xlsx"

draft_df = pd.read_excel(file_path, sheet_name="draft_and_player_data")
fantasy_df = pd.read_excel(file_path, sheet_name="fantasy_targets")
master_df = pd.read_excel(file_path, sheet_name="master_dataset")

print("Draft Data:", draft_df.shape)
print("Fantasy Targets:", fantasy_df.shape)
print("Master Dataset:", master_df.shape)
```

```python
# Preview Fantasy Target Table

fantasy_df.head()
```

```python
# Preview Master Table

master_df.head()
```

```python
# Check Column Names

print("Draft Columns:")
print(draft_df.columns.tolist())

print("\nFantasy Columns:")
print(fantasy_df.columns.tolist())

print("\nMaster Columns:")
print(master_df.columns.tolist())
```

```python
# Check Missing Values

print("Draft Missing Values:")
display(draft_df.isnull().sum())

print("Fantasy Missing Values:")
display(fantasy_df.isnull().sum())

print("Master Missing Values:")
display(master_df.isnull().sum())
```

```python
# Check data types

master_df.info()
```

```python
# Check Season Count

master_df["season"].value_counts().sort_index()
```

```python
# Create Train/Test/Predict Split

train_df = master_df[
    (master_df["season"] >= 2010) & 
    (master_df["season"] <= 2022)
].copy()

test_df = master_df[
    (master_df["season"] >= 2023) & 
    (master_df["season"] <= 2025)
].copy()

predict_2026_df = master_df[
    master_df["season"] == 2026
].copy()

print("Train:", train_df.shape)
print("Test:", test_df.shape)
print("Predict 2026:", predict_2026_df.shape)
```

```python
# Check Target Columns

target_regression = "rookie_total_ppr_points"
target_classification = "fantasy_relevant_binary"

print("Missing regression targets:")
print("Train:", train_df[target_regression].isnull().sum())
print("Test:", test_df[target_regression].isnull().sum())
print("Predict 2026:", predict_2026_df[target_regression].isnull().sum())

print("\nMissing classification targets:")
print("Train:", train_df[target_classification].isnull().sum())
print("Test:", test_df[target_classification].isnull().sum())
print("Predict 2026:", predict_2026_df[target_classification].isnull().sum())
```

```python
# Save Split Datasets

train_df.to_csv("data/processed/train_2010_2022.csv", index=False)
test_df.to_csv("data/processed/test_2023_2025.csv", index=False)
predict_2026_df.to_csv("data/processed/predict_2026.csv", index=False)

print("Files saved successfully.")
```
